## Save state and load state

```python
def save_state(dataframe, filename="app_state.csv"):
    dataframe.to_csv(filename, index=False)
    
def load_state(filename="app_state.csv"):
    return pd.read_csv(filename, parse_dates=["datetime"])
```

## Update with new rows

```python
def add_hour(df):
    new_row = df.iloc[-1].copy()                    
    new_row["datetime"] = new_row["datetime"] + pd.Timedelta(hours=1) 
    df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)   
    return df

def add_hours(df, n=1):
    for _ in range(n):
        df = add_hour(df)
    return df

df = add_hours(df, n = 5)
save_state(df)
```
