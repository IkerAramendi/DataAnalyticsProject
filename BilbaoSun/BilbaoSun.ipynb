```python
import numpy as np
import pandas as pd
df = pd.read_csv(
    "BilbaoSun.csv",
     skiprows=9
)
df.head(48)
```

## Rename columns

```python
df_renamed = df.rename(columns={
    "YEAR": "year",
    "MO": "month",
    "DY": "day",
    "HR": "hour"
})
```

## Create datetime column

```python
df_renamed["datetime"] = pd.to_datetime(df_renamed[["year","month","day","hour"]])
```

## Convert Wh/m² to kWh/m²

```python
df_renamed["irr_kWh_m2"] = df_renamed["ALLSKY_SFC_SW_DWN"] / 1000
```

## Max

```python
irradiacion_max_mensual = (
    df_renamed.groupby('month')['ALLSKY_SFC_SW_DWN']
      .max()
      .reset_index()
)

print(irradiacion_max_mensual)
```
